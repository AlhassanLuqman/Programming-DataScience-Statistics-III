```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

k = 0.3

def dydt(y, t):
    return -k * y

y0 = 5

t = np.linspace(0, 20, 200)

y = odeint(dydt, y0, t)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(t, y, label="$y(t)$")
plt.xlabel("Time t")
plt.ylabel("y(t)")
plt.title("ODE Solution of dy/dt = -k*y")
plt.legend()
plt.grid()
plt.show()
```
